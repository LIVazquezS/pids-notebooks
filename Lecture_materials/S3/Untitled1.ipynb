```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.plot(np.random.rand(10))
plt.legend(["Beautiful line 1", "Ugly line 2"])
```

```python
# Line plot

x = np.linspace(-5, 5, 20)
f1 = np.sin(x)
f2 = np.cos(x)
f3 = 1 / (1 + x**2)
```

```python
plt.plot(x, f1, 'ro-')
plt.plot(x, f2, 'k*--')
plt.plot(x, f3, 'gv-.')
plt.legend(["A sine function", "A cosine function", "The Witch of Agnesi"])
```

```python
# Scatter plots (you can interleave but then change order and everything)

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax2.plot(np.random.randn(10), 'g')
ax1.plot(np.random.randn(10))
ax1.set_title("A venerable title")
ax2.set_title("A venerable title")
ax3.set_xlabel('a label')
```
