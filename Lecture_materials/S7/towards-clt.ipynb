```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
import seaborn as sns
sns.set_style("whitegrid")
import pandas as pd
import random
import numpy as np
```

## Tossing a coin

```python
# (slow code; fast code example below)
import random

n_games = 10000
n_tosses = 100

tails_per_game = []
for i_game in range(n_games):
    n_tails = 0
    for i in range(n_tosses):
        toss = random.randint(0, 1)
        n_tails += 1 if toss == 0 else 0
    tails_per_game.append(n_tails)
        
sns.histplot(tails_per_game)
```

## Rolling a die

```python
import random

n_games = 10000
n_rolls = 100

# Distribution of the number of fours in a hundred rolls.
fours_per_game = []
for i_game in range(n_games):
    n_fours = 0
    for i in range(n_rolls):
        roll = random.randint(1, 6)
        n_fours += 1 if roll == 4 else 0
    fours_per_game.append(n_fours)
        
sns.histplot(fours_per_game)

# The average number of fours in a hundred rolls is about 16.
# That looks about right, since we "expect" to see about 100/6 = 16 + 2/3
# fours. But we see that in a 1000 games the number ranges between 
# as low as 5 and as high as 29
```

```python
# Distribution of outcomes of rolling one die (in theory)

die_outcomes = np.arange(1, 6 + 1)
p_die = np.ones((6,)) / 6
plt.stem(die_outcomes, p_die)
plt.title("Die probabilities")
plt.xlabel("Number rolled")
plt.ylabel("Probability")
```

```python
die_outcomes[np.newaxis, :] + die_outcomes[:, np.newaxis]
```

```python
# Distribution of outcomes of summation of two dice

two_dice_table = die_outcomes[:, None] + die_outcomes[None, :]
two_dice_outcomes = np.unique(two_dice_table)
p_two_dice = [(two_dice_table == outcome).sum() / 6**2 for outcome in two_dice_outcomes]
plt.stem(two_dice_outcomes, p_two_dice)
plt.title("Sum of two dice")
plt.xlabel("Sum of numbers rolled")
plt.ylabel("Probability of sum")
```

```python
# Distribution of outcomes of summation of three dice

three_dice_table = die_outcomes[:, None, None] + \
                   die_outcomes[None, :, None] + \
                   die_outcomes[None, None, :]
three_dice_outcomes = np.unique(three_dice_table)
p_three_dice = [(three_dice_table == outcome).sum() / 6**3 for outcome in three_dice_outcomes]
plt.stem(three_dice_outcomes, p_three_dice)
plt.title("Sum of three dice")
plt.xlabel("Sum of numbers rolled")
plt.ylabel("Probability of sum")
```
