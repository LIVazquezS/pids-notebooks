Check if this phenomenon appears only for Switzerland. 

```python
# Display the number of dentists over time for Europe
dentists_eu = dentists[dentists["ParentLocation"] == "Europe"]
dentists_eu = dentists_eu.groupby('Period').sum().reset_index()
dentists_eu.plot(x = 'Period' , y = 'FactValueNumeric')
```

Display the number of dentists for all continents. Do they follow the same trend?

```python
# Display the number of dentists over all continents on the same plots
import numpy as np
continent_names = np.unique(dentists["ParentLocation"])
print(continent_names)
dentists_continent = pd.DataFrame({})
cnt = 0
for continent in continent_names:
    cnt += 1
    print(continent)
    dentists_cont = dentists[dentists["ParentLocation"] == continent]
    dentists_cont = dentists_cont.groupby('Period').sum().reset_index()
    if cnt == 1:
        period_col = dentists_cont['Period']
        dentists_continent.insert(0, "Period", period_col)
        
    value_col = dentists_cont['FactValueNumeric']
    dentists_continent.insert(cnt, continent, value_col)

dentists_continent.head()


dentists_continent.plot(x = 'Period' , y = list(continent_names))
```

## 3g - Plots using Matplotlib

Let's now see what is happening in Switzerland. 


Get the Swiss population over the years.

```python
population_ch = population[population['Country Name']=='Switzerland']

ye = np.array(population_ch.columns[4:-1].values,dtype=int)
pop = population_ch[population_ch.columns[4:-1]].values[0]

print(ye)
```

Get the number of dentists over the years. 

```python
ye_dentists = dentists_ch['Period'].values
num_dentists = dentists_ch['FactValueNumeric'].values

print(ye_dentists)
```

Compute the ratio.

```python
#ratio = ... 
# TODO: do it, at least with for loop etc
ratio_val = np.zeros(ye_dentists.shape[0])
ratio_year = np.zeros(ye_dentists.shape[0])
for k in range(ye_dentists.shape[0]):
    nh = pop[ye==ye_dentists[k]]
    ratio_val[k] = nh/num_dentists[k]
    ratio_year[k] =  ye_dentists[k]
```

Plot the ratio.

```python
import matplotlib.pyplot as plt
plt.plot(ratio_year,ratio_val)
```
